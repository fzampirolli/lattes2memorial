```python
# Buscar arquivos .pdf no diretório atual e armazenar os caminhos em uma lista
import os
pdf_files = [f for f in os.listdir('.') if f.endswith('.pdf')]

# Exibir os arquivos encontrados
print(pdf_files)

# Verificar se há arquivos encontrados
if pdf_files:
    # Escolher o arquivo PDF desejado (por exemplo, o primeiro arquivo encontrado)
    pdf_file = pdf_files[0]  # Alterar conforme necessário

    # Fazer o download do arquivo encontrado
    from google.colab import files
    files.download(pdf_file)
else:
    print("Nenhum arquivo PDF encontrado.")
```

```python
import zipfile
import os

# Caminho da pasta que será compactada
folder_path = '../src'

# Caminho para o arquivo zip
zip_filename = '/content/src_folder.zip'

# Compactando a pasta em um arquivo zip
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    # Caminha por todos os arquivos e subpastas dentro da pasta '../src'
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # Caminho completo do arquivo
            file_path = os.path.join(root, file)
            # Adiciona o arquivo ao zip
            zipf.write(file_path, os.path.relpath(file_path, folder_path))

# Fazer o download do arquivo zip
from google.colab import files
files.download(zip_filename)
```
