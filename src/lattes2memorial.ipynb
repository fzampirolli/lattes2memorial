````python
import zipfile
import os

# Caminho da pasta que será compactada
folder_path = '../src'

# Caminho para o arquivo zip
zip_filename = '/content/src_folder.zip'

# Compactando a pasta em um arquivo zip
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    # Caminha por todos os arquivos e subpastas dentro da pasta '../src'
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # Caminho completo do arquivo
            file_path = os.path.join(root, file)
            # Adiciona o arquivo ao zip
            zipf.write(file_path, os.path.relpath(file_path, folder_path))

# Fazer o download do arquivo zip
from google.colab import files
files.download(zip_filename)
````

## Para incluir mais itens do Lattes no Memorial

**Objetivo:** Adicionar novos métodos à classe `lattes` para extrair informações do XML do Lattes e incluí-las no Memorial.

### Passo a passo (na pasta `src`)

1. **Abrir a classe `lattes`**

   * Local: final do arquivo `lattesClass.py`.

2. **Adicionar novo método**

   * Pedir ao ChatGPT (ou outra LLM) para adaptar um método da classe `lattes`. Métodos existentes:

     * Método com gráfico de barras.
     * Método com gráfico de barras e grafo de colaboradores.
     * Método sem gráficos.

   * Especifique qual **tag do XML** você quer processar (veja todas as tags no final do arquivo `lattesClass.py`).

3. **Chamar o método no script principal**

   * Arquivo: `lattes2memorial.py`.
   * Exemplo de uso:

     ```python
     ...
     lattes.novo_metodo()
     ...
     ```
   * Saída: arquivo `texLattes/novo_metodo.tex` e imagens `.png` em `figs`.
